## 2 - Jointure des dataframes

```python
df = impact.merge(qualite, on = 'code', how= 'inner', indicator = True)
df
```

```python
df.groupby(["brands_tags_x"])
df
```

```python
brands_not_showing_nutriscore = df[df["nutriscore_grade"].isnull()]["brands_tags_x"].drop_duplicates()
print("Marques ne montrant pas le nutriscore:")
print(brands_not_showing_nutriscore)
```

```python
brands_not_showing_ecoscore = df[df["ecoscore_grade"].isnull()]["brands_tags_x"].drop_duplicates()
print("Marques ne montrant pas l'Eco-Score:")
print(brands_not_showing_ecoscore)
```

```python

brands_with_nutriscore = df["nutriscore_grade"].notnull().sum()
total_brands = len(df["brands_tags_x"])
percentage_nutriscore = (brands_with_nutriscore / total_brands) * 100


brands_with_ecoscore = df["ecoscore_grade"].notnull().sum()
percentage_ecoscore = (brands_with_ecoscore / total_brands) * 100


brands_with_both_scores = df[(df["nutriscore_grade"].notnull()) & (df["ecoscore_grade"].notnull())]["brands_tags_x"].nunique()
percentage_both_scores = (brands_with_both_scores / total_brands) * 100

print("Pourcentage des marques qui affichent le Nutri-Score:", percentage_nutriscore)
print("Pourcentage des marques qui affichent l'Eco-Score:", percentage_ecoscore)
print("Pourcentage des marques qui affichent à la fois le Nutri-Score et l'Eco-Score:", percentage_both_scores)

```
